```python
import pandas as pd
from datetime import datetime, timezone, timedelta
```

# Lugares

```python
areaexp = pd.read_csv('../db/csv/areaexp.csv')
areapro = pd.read_csv('../db/csv/areapro.csv')
barrio = pd.read_csv('../db/csv/barrio.csv')
comuna = pd.read_csv('../db/csv/comuna.csv')
sector = pd.read_csv('../db/csv/sector.csv')
vereda = pd.read_csv('../db/csv/vereda.csv')
```

```python
created = datetime.now(tz=timezone(offset=-timedelta(hours=5), name='America/Bogota')).isoformat()
updated = datetime.now(tz=timezone(offset=-timedelta(hours=5), name='America/Bogota')).isoformat()
```

```python
vereda['nombre'] = vereda['nombre'].str.title()
vereda.head()
```

```python
# Lugares Codes
COMUNACODES = 1000
BARRIOCODES = 2000
VEREDACODES = 3000
SECTORCODES = 4000
AREAEXPCODES = 5000
AREAPROCODES = 6000
```

```python
areaexp['id'] = range(AREAEXPCODES, AREAEXPCODES + len(areaexp['id']))
areapro['id'] = range(AREAPROCODES, AREAPROCODES + len(areapro['id']))
barrio['id'] = range(BARRIOCODES, BARRIOCODES + len(barrio['id']))
comuna['id'] = range(COMUNACODES, COMUNACODES + len(comuna['id']))
sector['id'] = range(SECTORCODES, SECTORCODES + len(sector['id']))
vereda['id'] = range(VEREDACODES, VEREDACODES + len(vereda['id']))
```

```python
barrio['id_comuna'] = barrio['id_comuna'].apply(lambda id: (id - 1) + COMUNACODES if id < COMUNACODES else id)
```

```python
sector['id_vereda'] = sector['id_vereda'].apply(lambda id: (id - 1) + VEREDACODES if id < COMUNACODES else id)
```

```python
lugares = [lugar[['id', 'nombre']] for lugar in [comuna, barrio, vereda, sector, areaexp, areapro]]
lugares = pd.concat(lugares)
lugares['created'] = created
lugares['updated'] = updated
lugares
```

```python
lugares.to_csv('../db/csv/lugares.csv', index_label='id', index=False)
```
